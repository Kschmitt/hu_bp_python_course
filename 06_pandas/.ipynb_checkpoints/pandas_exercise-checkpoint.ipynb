### Task 9: Merge the two data sets (hint: "Kreisschlüssel" is the same as "AGS")

```python
merged = pd.merge(kern,keime,left_index=True, right_on="Kreisschlussel")
merged.head()

```

###  Task 10: Plot number of hospitalized patients per inhabitant against relative number of old people (65+)

```python
x= merged['Krankenhaus- Patienten 2013'].div(merged['EWZ'], axis=0)
y= merged['ALTER_5'].div(merged['EWZ'], axis=0)
```
